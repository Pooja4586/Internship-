```python
#importing the libraries needed 
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from time import sleep
from random import randint

#Declaring the headers 
headers = {"Accept-Language": "en-US,en;q=0.5"}

#declaring the list of empty variables, So that we can append the data overall
#creating an array of values and passing it in the url for dynamic webpages


#the whole core of the script
#creating an array of values and passing it in the url for dynamic webpages


metascore =[]
gross = []
#the whole core of the script

page = requests.get("https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1")
soup = BeautifulSoup(page.text, 'html.parser')
movie_data = soup.findAll('div', attrs = {'class': 'lister-item mode-advanced'})

for store in movie_data:
    meta = store.find('span', class_ = "metascore").text if store.find('span', class_ = "metascore") else "NA"
    metascore.append(meta)


    value = store.find_all('span', attrs = {'name': "nv"})


    grosses = value[1].text if len(value)>1 else 'NA'
    gross.append(grosses)

len(metascore)
```

```python
import selenium
import warnings
import pandas as pd
import time
from selenium.webdriver.common.action_chains import ActionChains
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

s=Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=s)
driver.maximize_window()

#linking Site

driver.get("https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1")

time.sleep(5)

Name=[]
Years=[]
Ratings=[]
Votes=[]
Genre1=[]
Genre2=[]
Genre3=[]
Runtime=[]
Metascore=[]
Director=[]
Certificate=[]
Gross=[]
Stars=[]

try:
    name=driver.find_elements(By.XPATH,"//*[@id='main']/div/div[3]/div/div/div[3]/h3/a")

    for i in name:
        Name.append(i.text)     #Scraping name list

except NoSuchElementException:
    Name.append("-")

try:
    span=driver.find_elements(By.XPATH,'//*[@id="main"]/div/div[3]/div/div/div[3]/h3/span[2]')

    for i in span:
        Years.append(i.text)          #Scraping Year span

except NoSuchElementException:
    Years.append("-")

try:
    genre=driver.find_elements(By.XPATH,'//span[@class="genre"]')

    for i in genre:
        x=i.text.split(",")
        if(len(x)==1):
            Genre1.append(x[0])
            Genre2.append("-")
            Genre3.append("-")
        elif(len(x)==2):
            Genre1.append(x[0])
            Genre2.append(x[1])
            Genre3.append("-")
        else:
            Genre1.append(x[0])
            Genre2.append(x[1])
            Genre3.append(x[2])

except NoSuchElementException:
    Genre.append("-")
try:
    runtime=driver.find_elements(By.XPATH,'//span[@class="runtime"]')
    for i in runtime:
        Runtime.append(i.text.strip("min"))      #Scraping Runtime

except NoSuchElementException:
    Runtime.append("-")
    
    
try:
    rating=driver.find_elements(By.XPATH,'//div[@class="inline-block ratings-imdb-rating"]')
    for i in rating:
        Ratings.append(i.text)            #Scraping Rating

except NoSuchElementException:
    Ratings.append("-")
    
try:
    vote=driver.find_elements(By.XPATH,'//*[@id="main"]/div/div[3]/div/div/div[3]/p[4]/span[2]')
    for i in vote:
        Votes.append(i.text)            #Scraping votes

except NoSuchElementException:
    Votes.append("-")
    


    

    
try:
    certt=driver.find_elements(By.XPATH,'//*[@id="main"]/div/div[3]/div/div/div[3]/p[1]/span[1]')
    
    for i in certt:
        Certificate.append(i.text)            #Scraping Certificate

except NoSuchElementException:
    Certificate.append("-")
    


Cast=[]
Dir1=[]
Dir2=[]
Dir3=[]
Actor1=[]
Actor2=[]
Actor3=[]
Actor4=[]
Cast1=[]
Cast2=[]

try:
    actor=driver.find_elements(By.XPATH,'//*[@id="main"]/div/div[3]/div/div/div[3]/p[3]')
    
    for i in actor:
        y=i.text.split("|")
        Cast1.append(y[0])
        Cast2.append(y[1])
        
        
                                       #Scraping Stars

except NoSuchElementException:
    Cast.append("-")
    
n=len(Cast1)
for i in range(0,n):
    str1=Cast1[i]
    x=str1.split(",")
    if(len(x)==1):
        temp=x[0]
        y=temp.split(":")
        Dir1.append(y[1])
        Dir2.append("-")
        Dir3.append("-")
    elif(len(x)==2):
        temp=x[0]
        y=temp.split(":")
        Dir1.append(y[1])
        Dir2.append(x[1])
        Dir3.append("-")
    else:
        temp=x[0]
        y=temp.split(":")
        Dir1.append(y[1])
        Dir2.append(x[1])
        Dir3.append(x[2])

k=len(Cast2)
for i in range(0,k):
    str1=Cast2[i]
    x=str1.split(",")
    if(len(x)==1):
        temp=x[0]
        y=temp.split(":")
        Actor1.append(y[1])
        Actor2.append("-")
        Actor3.append("-")
        Actor4.append("-")
    elif(len(x)==2):
        temp=x[0]
        y=temp.split(":")
        Actor1.append(y[1])
        Actor2.append(x[1])
        Actor3.append("-")
        Actor4.append("-") 
    elif(len(x)==3):
        temp=x[0]
        y=temp.split(":")
        Actor1.append(y[1])
        Actor2.append(x[1])
        Actor3.append(x[2])
        Actor4.append("-")
    else:
        temp=x[0]
        y=temp.split(":")
        Actor1.append(y[1])
        Actor2.append(x[1])
        Actor3.append(x[2])
        Actor4.append(x[3])
    
df1=pd.DataFrame({"Name":Name,"Dir1":Dir1,"Dir2":Dir2,
                  "Dir3":Dir3,"Years":Years,
                 "Runtime":Runtime,"Ratings":Ratings,"metascore":metascore})
                  
df2=pd.DataFrame({"Name":Name,"Actor1":Actor1,"Actor2":Actor2,"Actor3":Actor3,
                  "Actor4":Actor4,"Genre1":Genre1,"Genre2":Genre2,
                  "Genre3":Genre3,"Votes":Votes,
                 "Certificate" : Certificate,"gross":gross})




#Movie Name, Director Name, Duration, year, ratings, Metascore

#Movie Name, stars, votes, Genre, Gross collection, popularity, Certification
```
